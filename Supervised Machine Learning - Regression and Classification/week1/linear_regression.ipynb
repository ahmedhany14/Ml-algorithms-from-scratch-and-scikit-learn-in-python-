```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as nb
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split # used to split data set into (x, y) train and (x, y) test 
from sklearn.preprocessing import StandardScaler # used to transform data in to stadarization form
from sklearn.metrics import accuracy_score
df = pd.read_csv(r'/home/ahmed/Ai/Analysis/Data analysies/scikt learn/world_population.csv')
df
```

# spliting data to train and test

```python
train, test = df[['high_GPA', 'math_SAT', 'verb_SAT', 'comp_GPA']], df['univ_GPA']
x_train, x_test, y_train, y_test = train_test_split(train, test, test_size=.2,random_state=0)
```

# defining an object from LinearRegression, and then train the model

### fit
fit function used to train the model, taks 2 parametars (inputs, outputs)
                                                         
    x_train, y_train

```python
lin_reg = LinearRegression()
lin_reg.fit(X=x_train,y=y_train)
```

### predict
predict function used to get predicted results from model, which taks 1 parametars (inputs)
                                                         
    x_test

return output

    predict of model

```python
predict = lin_reg.predict(x_test)
print('The result of model is : ')
print(predict)
```

```python
samples = len(x_test)
accual = list(y_test)
for i, j in zip(accual, predict):
    print('accual = ', i, '  --->     predict = ', j)
```

### score
score function used to compare the result of model output and accual output, which taks 1 parametars (inputs)
                                                         
    x_test, y_test

return percentage

    score of model

```python
print('The socre of the model = ', lin_reg.score(X=x_test,y=y_test) * 100, '%')
```
