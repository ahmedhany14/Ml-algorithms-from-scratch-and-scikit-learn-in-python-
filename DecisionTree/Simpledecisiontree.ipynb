```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, plot_tree
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv(
    r"/home/ahmed/Ai/Ml-algorithms-from-scratch-and-scikit-learn-in-python-/DataSets/salaries.csv"
)
df
```

```python
l = LabelEncoder()

df["company"] = l.fit_transform(df["company"])
df["job"] = l.fit_transform(df["job"])
df["degree"] = l.fit_transform(df["degree"])
```

```python
Y = df["salary_more_then_100k"]
df = df.drop(["salary_more_then_100k"], axis="columns")
```

```python
model = DecisionTreeClassifier(
    max_depth=10,
    criterion="gini",
    min_samples_leaf=2,
    min_samples_split=4,
    min_impurity_decrease=0.093,
)
model.fit(X=df, y=Y)
```

```python
model.score(df, Y) * 100
```

```python
pre = model.predict(df)
list(pre), list(Y)
```

```python
plot_tree(model)
plt.show()
```
