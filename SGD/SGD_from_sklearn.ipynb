```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import (
    SGDClassifier,
    SGDRegressor,
    LinearRegression,
    LogisticRegression,
)
from sklearn.datasets import make_regression, make_classification
import time

x_train, y_train = make_regression(
    n_samples=2000,
    n_features=1,
    n_informative=1,
    n_targets=1,
    noise=2,
    random_state=10,
)
x_test, y_test = make_regression(
    n_samples=2000, n_features=1, n_informative=1, n_targets=1, noise=1, random_state=10
)
```

# SGD

SGD has been successfully applied to large-scale and sparse machine learning problems often encountered in text classification and natural language processing. Given that the data is sparse, the classifiers in this module easily scale to problems with more than 10^5 training examples and more than 10^5 features.

The advantages of Stochastic Gradient Descent are:

- Efficiency.

- Easy of implementation (lots of opportunities for code tuning).

The disadvantages of Stochastic Gradient Descent include:

- SGD requires a number of hyperparameters such as the regularization parameter and the number of iterations.

- SGD is sensitive to feature scaling.

SGD Complexity: O(iter _ n_features _ m_samples)


# SGD Regression


```python
SGD = SGDRegressor(
    loss="squared_error",
    penalty="elasticnet",
    alpha=0.00000001,
    l1_ratio=0.05,
    max_iter=10000000,
    learning_rate="invscaling",
)
SGD.fit(x_train, y_train)
```

```python
SGD.coef_, SGD.intercept_, SGD.score(x_train, y_train), SGD.score(x_test, y_test)
```

# SGD Classifier


```python
x_train, y_train = make_classification(
    n_samples=10000,
    n_classes=2,
    random_state=10,
)
x_test, y_test = make_classification(n_samples=100, n_classes=2, random_state=10)
```

```python
SGD = SGDClassifier(
    loss="modified_huber",
    penalty="elasticnet",
    alpha=0.0001,
    l1_ratio=0.005,
    max_iter=10000000,
    learning_rate="invscaling",
    eta0=1,
)
SGD.fit(x_train, y_train)
```

```python
SGD.coef_, SGD.intercept_, SGD.score(x_train, y_train), SGD.score(x_test, y_test)
```
