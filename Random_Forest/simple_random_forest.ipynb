```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import OneHotEncoder

df = sns.load_dataset("penguins")
df.dropna(inplace=True)
```

```python
df_encoded = pd.get_dummies(
    df, columns=["sex", "island"], drop_first=True, dtype="int32"
)
df_encoded
```

```python
features = df_encoded.drop(columns="species", axis="columns")
mapper = {"Adelie": 0, "Gentoo": 1, "Chinstrap": 2}
test = df_encoded["species"].replace(mapper)
x_train, x_test, y_train, y_test = train_test_split(
    features, test, test_size=0.2, random_state=42
)
```

```python
RF = RandomForestClassifier(criterion="gini", n_estimators=5, random_state=0)
RF.fit(x_train, y_train)
print(RF.score(x_train, y_train)), print(RF.score(x_test, y_test))
```
