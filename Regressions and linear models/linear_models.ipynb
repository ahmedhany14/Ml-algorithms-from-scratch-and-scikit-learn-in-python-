```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.model_selection import (
    train_test_split,
)  # used to split data set into (x, y) train and (x, y) test
from sklearn.preprocessing import (
    StandardScaler,
)  # used to transform data in to stadarization form
from sklearn.datasets import make_regression
from sklearn.metrics import accuracy_score

pd.options.display.max_columns = None
x_train, y_train = make_regression(
    n_samples=1000,
    n_features=1,
    n_informative=1,
    n_targets=1,
    noise=10,
    random_state=10,
)
x_test, y_test = make_regression(
    n_samples=100, n_features=1, n_informative=1, n_targets=1, noise=10, random_state=10
)
```

```python
plt.scatter(x=x_train, y=y_train)
plt.title("Train data")
```

```python
plt.scatter(x=x_test, y=y_test)
plt.title("test data")
```

# linear regression


```python
lin_model = LinearRegression()
lin_model.fit(X=x_train, y=y_train)
print("linear model coef = ", lin_model.coef_)
print("linear model intercept = ", lin_model.intercept_)
```

```python
pred = lin_model.predict(X=x_test)
score_of_linear = lin_model.score(X=x_test, y=y_test)
print("model score = ", score_of_linear * 100, "%")
```

```python
plt.scatter(x=x_train, y=y_train, label='accual"')
plt.plot(x_test, pred, color="r", label="prediction")
plt.legend()
plt.show()
```

# Lasso

The diffrence between lasso and linear, that ridge used to increase or decrease slope based on adding this term to linear model

    alpha * W


```python
print("linear model score = ", score_of_linear * 100, "%")
alpha = [0, 5, 8, 12, 15, 19, 26, 30, 35]
plt.figure(figsize=(12, 6))
plt.scatter(x=x_train, y=y_train, label="accual")

plt.scatter(
    x=x_test,
    y=y_test,
    label="Test",
    color="black",
)
plt.title("test data")
max_score_so_far = score_of_linear
coef, intercept = lin_model.coef_, lin_model.intercept_
best_model = "linear regression"

for i in alpha:
    lasso_reg = Lasso(alpha=i)
    lasso_reg.fit(X=x_train, y=y_train)
    pred = lasso_reg.predict(X=x_test)
    lasso_score = lasso_reg.score(X=x_test, y=y_test)
    plt.plot(x_test, pred, label="prediction when alpha = " + str(i))
    print("lasso model score when alpta = ", i, " = ", lasso_score * 100, "%")
    if lasso_score > max_score_so_far:
        max_score_so_far = lasso_score
        coef, intercept = lasso_reg.coef_, lasso_reg.intercept_
        best_model = "Lasso"
plt.legend()
plt.show()
```

```python
print("the best model till now is= ", best_model)
print("the best socre till now = ", max_score_so_far * 100, "%")
print("the best coef till now = ", coef)
print("the best intercept till now = ", intercept)
```

# Ridge

The diffrence between ridge and linear, that ridge used to increase or decrease slope based on adding this term to linear model

    alpha * W^2


```python
print("linear model score = ", score_of_linear * 100, "%")
alpha = [
    0,
    50,
    80,
    120,
    150,
    190,
    260,
    300,
    350,
    400,
    450,
    490,
    491,
    492,
    493,
    494,
    495,
    496,
    497,
    498,
    499,
    500,
    550,
    600,
]
plt.figure(figsize=(20, 10))
plt.scatter(x=x_train, y=y_train, label="accual")

plt.scatter(
    x=x_test,
    y=y_test,
    label="Test",
    color="black",
)
plt.title("test data")

for i in alpha:
    Ridge_reg = Ridge(alpha=i)
    Ridge_reg.fit(X=x_train, y=y_train)
    pred = Ridge_reg.predict(X=x_test)
    Ridge_score = Ridge_reg.score(X=x_test, y=y_test)
    plt.plot(x_test, pred, label="prediction when alpha = " + str(i))
    print("Ridge model score when alpta = ", i, " = ", Ridge_score * 100, "%")
    if Ridge_score > max_score_so_far:
        max_score_so_far = Ridge_score
        coef, intercept = Ridge_reg.coef_, Ridge_reg.intercept_
        best_model = "Ridge"
plt.legend()
plt.show()
```

```python
print("the best model till now is= ", best_model)
print("the best socre till now = ", max_score_so_far * 100, "%")
print("the best coef till now = ", coef)
print("the best intercept till now = ", intercept)
```

# Where is diffrent between Lasso and Ridge???

### Ridge

acctuly the main core of Ridge is to reduse or imposing a penalty on the size of the coefficients, by minimize residual sum of squares + alpha \* w ^ 2

### Lasso

The main core of Lasso is not to reduse the size of the coefficients, but reducing the number of features, so it recommendly to use it with models that have many features, but there is some features has less piriority from others, so Lasso works to make their coefficients near to 0


# Elastic-Net

Elastic-Net conseere a combination of Lasso and ridge, which works on reduseing the coefficients and reduse unneeded features

    alpha * w + (alpha(1 - lambda) / 2) * w^2


```python
print("linear model score = ", score_of_linear * 100, "%")
alpha = [0, 0.52, 1, 1.5]
lambda_ = [0, 0.05, 0.02, 0.03, 0.04]
plt.figure(figsize=(20, 10))
plt.scatter(x=x_train, y=y_train, label="accual")

plt.scatter(
    x=x_test,
    y=y_test,
    label="Test",
    color="black",
)
plt.title("test data")

for i in alpha:
    for j in lambda_:
        ElasticNet_reg = ElasticNet(alpha=i, l1_ratio=j)
        ElasticNet_reg.fit(X=x_train, y=y_train)
        pred = ElasticNet_reg.predict(X=x_test)
        ElasticNet_score = ElasticNet_reg.score(X=x_test, y=y_test)
        plt.plot(
            x_test,
            pred,
            label="prediction when alpha = " + str(i) + "and lambda = " + str(j),
        )
        print(
            "ElasticNet model score when alpha = ",
            i,
            " and lambda = ",
            j,
            " = ",
            ElasticNet_score * 100,
            "%",
        )
        if ElasticNet_score > max_score_so_far:
            max_score_so_far = ElasticNet_score
            coef, intercept = ElasticNet_reg.coef_, ElasticNet_reg.intercept_
            best_model = "ElasticNet"
plt.legend()
plt.show()
```

```python
print("the best model till now is= ", best_model)
print("the best socre till now = ", max_score_so_far * 100, "%")
print("the best coef till now = ", coef)
print("the best intercept till now = ", intercept)
```
