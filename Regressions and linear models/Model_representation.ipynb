# Problem statment

Given a simple data set, with 2 features size and price of house
fit a linear regressinn modle that pridict price for other houses

| Size (1000 sqft) | Price (1000s of dollars) |
| ---------------- | ------------------------ |
| 1.0              | 300                      |
| 2.0              | 500                      |


```python
# x_train -> size
# y_train -> price
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

print("x_train", x_train)
print("y_train", y_train)
```

```python
m = x_train.shape[0]
print("size of simples = ", m)
```

# plotting the traing data set


```python
plt.scatter(
    x=x_train,
    y=y_train,
    color="red",
    marker=".",
)

plt.title("Houses prices")
plt.xlabel("House size")
plt.ylabel("House price")

plt.show()
```

# Single feature linear regression model

$$ f\_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{1}$$


```python
# adusting the model parameters values (w, b)
w = 100
b = 100
print("parameter w = ", w)
print("parameter b = ", b)
```

# creating function model

Now we need to compute the value of $f_{w,b}(x^{(i)})$ for each point in data set


```python
def compute_model_fucntion(x_train, w, b):
    f_wb = [0] * m
    for i in range(m):
        f_wb[i] = w * x_train[i] + b
    return f_wb
```

```python
f_wb = compute_model_fucntion(x_train, w, b)
f_wb
```

# plotting the result of model function and comparing it with the real result


```python
plt.scatter(x=x_train, y=y_train, color="red", marker=".", label="Actual Values")

plt.plot(x_train, f_wb, color="blue", label="Predict values")


plt.title("Houses prices")
plt.xlabel("House size")
plt.ylabel("House price")

plt.show()
```

# prediction


```python
def predict(w, b, x_i):
    f_wb = w * x_i + b
    return f_wb
```

```python
w = 200
b = 100
size = 1.2

cost_of_house = predict(w, b, size)
print("cost of house with size 1.2 = ", cost_of_house)
```
